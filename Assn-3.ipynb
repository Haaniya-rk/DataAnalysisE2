```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

a = np.load( 'data-fig1.npy' )
plt.plot( a, 
         marker='o',
         color="black", linewidth=2  )
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

## Figure 2

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

a = np.load( 'data-fig2.npy' )
colors = ['red', 'green', 'blue']
labels = ['Group1', 'Group2', 'Group3']

for i in range(3):
    plt.plot( a[:, i], color=colors[i], label=labels[i],
         linewidth=2  )
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.show()
```

```python
%matplotlib inline

a = np.load( 'data-fig3.npy' )
x = a[:, 0]
y = a[:, 1]
plt.scatter( x, y, color='b', alpha=0.3, s=200)

plt.show()
```

## Figure 4


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

a = np.load( 'data-fig4.npy' )
xlabels = ['A', 'B', 'C', 'D', 'E']

plt.figure()

x = np.arange(len(a))

plt.bar(x, a, color='grey')

plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks( ticks=x, labels=xlabels)

plt.show()
```
